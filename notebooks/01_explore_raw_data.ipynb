```python
artists = pd.read_csv("../data/raw/artists.csv")
tracks = pd.read_csv("../data/raw/tracks.csv")
listening_events = pd.read_csv("../data/raw/listening_events.csv")
subscriptions = pd.read_csv("../data/raw/subscriptions.csv")
payments = pd.read_csv("../data/raw/payments.csv")

print("artists:", artists.shape)
print("tracks:", tracks.shape)
print("listening_events:", listening_events.shape)
print("subscriptions:", subscriptions.shape)
print("payments:", payments.shape)
```

```python
payments['amount'].sum()
successful = payments[payments["status"] == "success"]
successful.head()
```

```python
successful['month'] = pd.to_datetime(successful['payment_date']).dt.to_period('M')
mrr = successful.groupby('month')['amount'].sum()
mrr
```
